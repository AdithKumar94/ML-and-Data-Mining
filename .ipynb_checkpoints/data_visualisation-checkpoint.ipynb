```python
import pandas as pd
import numpy as np
import time
import math
import re
import matplotlib as mat

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

#read data from file
ratings = pd.read_csv("./ml-100k/u1.base",sep="\t",names = ["user_id", "item_id" , "ratings", "timestamp"])

users=pd.read_csv("./ml-100k/u.user",sep='|',names=['user_id', 'age', 'sex', 'occupation', 'zip_code'])
ratings_movies = ratings["item_id"];

genres = pd.read_csv("./ml-100k/u.item",sep="|",names = ["item_id", "movie_title",\
            "release date", "video release date", "IMDb URL", "unknown", "Action",\
            "Adventure","Animation", "Children's", "Comedy", "Crime", "Documentary",\
            "Drama","Fantasy", "Film-Noir", "Horror", "Musical", "Mystery",\
            "Romance","Sci-Fi", "Thriller", "War", "Western"], encoding='latin-1')


MovieG = list(genres.columns.values)[-19:]
movgen=[]
genres1 = np.array(genres)
for i in range(len(genres)):
    gen = []
    for j in range(6,24):
        if genres1[i][j] == 1:
            gen.append(MovieG[j - 5])
    movgen.append([i+1,gen])

for i in range(len(movgen)):
    movgen[i][1] = '|'.join(str(x) for x in movgen[i][1])
    


movgenDF = pd.DataFrame(movgen, columns=["item_id","genres"])    


final_genres = pd.merge(genres, movgenDF, on = "item_id")
final_genres = final_genres[["item_id","movie_title","genres"]]





def explode(df, lst_cols, fill_value='', preserve_index=False):
    # make sure `lst_cols` is list-alike
    if (lst_cols is not None
        and len(lst_cols) > 0
        and not isinstance(lst_cols, (list, tuple, np.ndarray, pd.Series))):
        lst_cols = [lst_cols]
    # all columns except `lst_cols`
    idx_cols = df.columns.difference(lst_cols)
    # calculate lengths of lists
    lens = df[lst_cols[0]].str.len()
    # preserve original index values    
    idx = np.repeat(df.index.values, lens)
    # create "exploded" DF
    res = (pd.DataFrame({
                col:np.repeat(df[col].values, lens)
                for col in idx_cols},
                index=idx)
             .assign(**{col:np.concatenate(df.loc[lens>0, col].values)
                            for col in lst_cols}))
    # append those rows that have empty lists
    if (lens == 0).any():
        # at least one list in cells is empty
        res = (res.append(df.loc[lens==0, idx_cols], sort=False)
                  .fillna(fill_value))
    # revert the original index order
    res = res.sort_index()
    # reset index if requested
    if not preserve_index:        
        res = res.reset_index(drop=True)
    return res


final_genres.genres = final_genres.genres.str.split('|')
final_genres = explode(final_genres, ["genres"])
final_genres["Movie Name"] = final_genres["movie_title"].str.split(" ").apply(lambda x: ' '.join(str(x) for x in x[:-1]))
final_genres["Year"] = final_genres["movie_title"].str.split(" ").apply(lambda x: x[-1].lstrip('(').rstrip(')'))
final_genres = final_genres[["item_id", "Movie Name", "Year", "genres"]]
print(final_genres.head(5))    

  

#ratings_dis = ratings.merge(final_genres, left_on='item_id', right_on='item_id', how='inner')




###################top 25#################
movielens=pd.merge(users,ratings)
movielens=pd.merge(movielens,genres)
#print(movielens.head(3))
mat.pyplot.figure(figsize=(20,6))
most_rated=movielens.groupby('movie_title').size().sort_values(ascending=False)[:10]
most_rated.plot(kind="bar",title="best rated 10 movies",label="count")



```

```python
##########################################
import seaborn as sns
import time
import matplotlib as mat
%matplotlib inline

ratings['timestamp'] = ratings['timestamp'].apply(lambda x: time.strftime('%Y', time.localtime(x)))

ratings = ratings.merge(final_genres, left_on='item_id', right_on='item_id', how='inner')

ratings2 = ratings.groupby(['item_id','timestamp','genres'], as_index=False)['ratings'].aggregate(np.mean).sort_values(by='ratings',ascending=False)
#ratings2=ratings['ratings'].size().sort_values(ascending=False)[:25]

mat.pyplot.figure(figsize=(60,30))
ax = sns.violinplot(x="timestamp",y="ratings",hue="genres",data=ratings2).set_title="Average Trends of rating in year for 2 years"
#plt.legend
```

```python
#Count by genres
%matplotlib inline
count=final_genres.groupby('genres').size().reset_index(name='counts')
mat.pyplot.figure(figsize=(20,10))
sns.barplot(x="genres",y="counts",data=count).set_title('Counts by genres')
```

```python
#count by years
%matplotlib inline
count=final_genres.groupby('Year').size().reset_index(name='counts')
mat.pyplot.figure(figsize=(30,10))
sns.barplot(x="Year",y="counts",data=count)
fnt =sns.barplot(x="Year",y="counts",data=count)
fnt.axes.set_title("Counts by Years",fontsize=50)
fnt.set_xlabel("Year",fontsize=30)
fnt.set_ylabel("counts",fontsize=50)


```
